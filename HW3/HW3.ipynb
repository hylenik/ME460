```python
### Problem 1 

from math import *
from sympy import * 

t1, t4, t5, t7, t8, r8 = symbols("t1 t4 t5 t7 t8 r8", real=True)
t2 = radians(40)
t3 = radians(118)
t9 = radians(357)


R1 = 80*exp(I*t1)
R2 = 62*exp(I*t2)
R3 = 61*exp(I*t3)
R4 = 73*exp(I*t4)
R5 = 39*exp(I*t5)
R6 = 43*exp(I*(t5-radians(2)))
R7 = 69*exp(I*t7)
R8 = r8*exp(I*t8)
R9 = 42*exp(I*t9)

eq1 = R2 + R3 + R4 -R5 - R1
eq2 = R5 + R6 + R7 - R8 
eq3 = R2 + R9 - R8 - R1

ans = nsolve((re(eq1),im(eq1),re(eq2),im(eq2),re(eq3),im(eq3)), (t1, t4, t5, t7, t8, r8), (radians(17),radians(330),radians(80),radians(265),radians(55),20), 
             maxsteps=100, verbose=false, prec=20, tol=0.02, verify=true, solver='anewton')

print("For input t3 = 118\N{DEGREE SIGN}: \n")
print("t1=",degrees(ans[0]),"\nt4=",degrees(ans[1]),"\nt5=",degrees(ans[2]),"\nt7=",degrees(ans[3]),'\nt8=',degrees(ans[4]),'\nr8=',ans[5])



```

```python
### Problem 2 

from math import *
from sympy import *
import numpy as np
import matplotlib.pyplot as plt

t1, t2, t5, t6, t7 = symbols("t1 t2 t5 t6 t7", real=True)
t10 = radians(245)
t11 = radians(250)

R1 = 97*exp(I*t1)
R2 = 57*exp(I*t2)
R3 = 61*exp(I*(t2 + radians(160)))
R4 = 21*exp(I*(t2 + radians(270)))
R5 = 24*exp(I*t5)
R6 = 60*exp(I*t6)
R7 = 55*exp(I*t7)
R8 = 79*exp(I*(t7 + radians(148)))
R9 = 44*exp(I*(t7 + radians(287)))
R10 = 98*exp(I*t10)
R11 = 16*exp(I*t11)

eq1 = R1 + R2 - R5 + R10
eq2 = R5 + R3 + R7 + R6 + R11
plotpt = R1 - R4 - R9

xpoint = np.zeros(360)
ypoint = np.zeros(360)

index = 0
for i in range(0,360,5):
    eq1s = eq1.subs(t5, radians(i))
    eq2s = eq2.subs(t5, radians(i))
    ans = nsolve((re(eq1s),im(eq1s),re(eq2s),im(eq2s)),(t1, t2, t6, t7), (radians(85), radians(300), radians(250), radians(40)),
                maxsteps=100, verbose=false, prec=20, tol=0.02, verify=false, solver='anewton')

    xpoint[index] = re(plotpt.subs(t1,ans[0]).subs(t2,ans[1]).subs(t7,ans[3]))
    ypoint[index] = im(plotpt.subs(t1,ans[0]).subs(t2,ans[1]).subs(t7,ans[3]))
    index = index + 1
    
plt.scatter(xpoint[:index], ypoint[:index])
plt.axes().set_aspect('equal','datalim')
plt.show()
```

```python
### Problem 3 

from math import *
from sympy import *
import numpy as np
from graphics import *
import matplotlib.pyplot as plt 

t2, t3, t4 = symbols("t2 t3 t4", real=True)
R1 = 49*exp(I*0)
R2 = 91*exp(I*t2)
R3 = 65*exp(I*t3)
R4 = 72*exp(I*t4)
eq = R2 + R3 - R4 - R1 

win = GraphWin("Ladder Swivel",640,640)
win.setCoords(-100,-100,200,200)

ground1 = Point(0,0)
ground2 = Point(re(R1),im(R1))

for i in range(15,150,5):
    eqsub = eq.subs(t4,radians(i))
    ans = nsolve((re(eqsub),im(eqsub)),(t2,t3),(radians(80),radians(70)))
    
    
    vect2 = R2.subs(t2,ans[0])
    vect3 = R3.subs(t3,ans[1])
    vect4 = R4.subs(t4,radians(i))
    
    v2pts = Point(re(vect2),im(vect2))
    v3pts = Point(re(vect2+vect3),im(vect2+vect3))
    v4pts = Point(re(R1+vect4),im(R1+vect4))
    
    Line(ground1,v2pts).draw(win)
    Line(v2pts,v3pts).draw(win)
    Line(ground2,v4pts).draw(win).setOutline('red')
    Line(ground1,ground2).draw(win).setOutline('green')
    
win.getMouse()
win.close()


    
```
